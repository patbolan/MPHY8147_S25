# Magnetic Dipole Field Lines

The prompt was to provide python code to plot the magnetic field lines from a magnetic dipole.
Small hand modifications by PJB to tweak appearance.

```python
import numpy as np
import matplotlib.pyplot as plt

def magnetic_dipole_field(x, y, m):
    r = np.sqrt(x**2 + y**2)
    r_cubed = r**3
    Bx = 3 * x * y * m / r_cubed
    By = (3 * y**2 * m - m * r**2) / r_cubed
    return Bx, By

Lx = 10

# Create a grid of points
x = np.linspace(-Lx, Lx, 200)
y = np.linspace(-Lx, Lx, 200)
X, Y = np.meshgrid(x, y)

# Calculate the magnetic field components
m = 1  # Dipole moment
Bx, By = magnetic_dipole_field(X, Y, m)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 10))
ax.streamplot(X, Y, Bx, By, density=4, color='b', linewidth=1, arrowsize=1.5)

# Add a circle to represent the dipole
circle = plt.Circle((0, 0), 0.2, color='r')
ax.add_artist(circle)

# Set plot limits and labels
ax.set_xlim(-Lx, Lx)
ax.set_ylim(-Lx, Lx)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Magnetic Field Lines of a Dipole')
ax.set_aspect('equal')

plt.show()
```
